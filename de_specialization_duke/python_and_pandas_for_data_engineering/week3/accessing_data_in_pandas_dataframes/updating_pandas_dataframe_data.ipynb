# Updating DataFrame Data

```python
import pandas as pd
data = {
    'first': ['Carl', 'Francies', 'Sam'],
    'last': ['Po', 'Nyguen', 'Smith'],
    'age': ['32', '45', '22']
}

clients = pd.DataFrame(data)
clients
```

# Adding Rows

```python
new_data = {
    'first': ['Sue', 'Boya'],
    'last': ['Rankler', 'Maple'],
    'age': ['93', '12'],
    'CH_count': [22, 1]
}

new_clients = pd.DataFrame(new_data)
new_clients
```

```python
clients = pd.concat([clients, new_clients])
clients
```

```python
clients.reset_index(inplace=True)
clients
```

# Setting Specific Value

```python
clients.loc[1, 'first'] = 'Frankie'
clients
```

```python
clients.loc[0:1, 'CH_count'] = -1
clients
```

# Math Operations

```python
# Not in place
clients.CH_count + 1
```

```python
# In place
clients.CH_count -= 3
clients
```

# Replace

```python
clients.replace(-7, 0, inplace=True)
clients
```
